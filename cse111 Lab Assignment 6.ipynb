```python
class SultansDine:
  num_branch=0
  sell=0
  branches=[]
  def __init__(self,branch):
    self.branch=branch
    SultansDine.num_branch+=1
    SultansDine.branches.append(self)
    self.quantity=0
    self.sell=SultansDine.sell

  def branchInformation(self):
    print('Branch Name:',self.branch)
    print('Branch Sell:',self.sell)

  def sellQuantity(self,quan):
    if quan<10:
      self.sell=quan*300
    elif quan<20:
      self.sell=quan*350
    else:
      self.sell=quan*400

  @classmethod
  def details(cls):
    Total_sell=0
    for i in cls.branches:
      Total_sell+=i.sell 
    print('Total Number of branch(s):',cls.num_branch)
    print('Total Sell:',Total_sell)
    for i in cls.branches:
      print(f'Branch Name: {i.branch},Branch Sell: {i.sell}')
      print(f"Branch consists of total sell's: {round((i.sell/Total_sell)*100,2)}%")

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
